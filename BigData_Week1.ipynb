### **Python Fundamentals**
Python is an **interpreted** programming language.
It can be used to develop programs based on procedural and object-oriented programming paradigms.
Python has a very simple syntax.
It provides many built-in programming features.
Primitive Types
There are several primitive types in python. Primitive type means a built-in type which is a part of the language. In short, python has some built-in data types and data structures which ease the overall coding experience. The most important ones are the followings:
>1. Integer (int)
2. Floating Point (float)
3. Boolean Logic (bool)
4. String (str)
5. List ('list' or can be denoted as square brackets)
6. Tuple ('tuple' or can be denoted as paranthesis)
7. Dictionary ('dict' or can be denoted as curly braces)

**Integer:** numeric values without fractional part. Examples: 1, 32, 999

**Floating Point:** numeric values with fractional part. Examples: 1.3, 3.2, 8.3454

**boolean:** logical values. Either true or false.

**string:** an ordered sequence of alphanumeric characters. Examples: 'Wolves', 'Python'
**list:** an ordered sequence of objects in which each element is identifed by an integer subscript. Examples: [], [1, 3, 4], ['a', 3, 7.5]

tuple: similar to the list type, but the content cannot be changed once it is created. Examples: (), (1, 3, 4), ('a', 3, 7.5)

**dictionary:** a collection of objects in which each object contains a key and a value. Examples: {}, {1: 'orange', 2: 'banana'}

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Correct URL (space replaced with %20)
url = "https://admin.opendatanepal.com/api/action/datastore_search?resource_id=b144c376-958a-4b3e-8126-e4b43ead5df0&sort=_id%20asc"

# Fetch and load data
response = requests.get(url)
data = response.json()
df = pd.DataFrame(data['result']['records'])

print("Initial data shape:", df.shape)
print("\nFirst 3 rows:")
print(df.head(3))
print("\nData types:")
print(df.dtypes)
```

```python
# Function to clean price columns
def clean_price(price_str):
    if isinstance(price_str, str):
        # Remove commas and convert to integer/float
        return float(price_str.replace(',', ''))
    return price_str

# Apply cleaning to the three price columns
price_columns = ['DAP Price', 'UREA Price', 'MOP Price']
for col in price_columns:
    df[col + '_clean'] = df[col].apply(clean_price)

# Drop original text columns and keep cleaned ones
df_clean = df.drop(columns=price_columns)
df_clean.columns = ['_id', 'Zone', 'DAP', 'UREA', 'MOP']

print("Cleaned data (first 5 rows):")
print(df_clean.head())
print("\nNew data types:")
print(df_clean.dtypes)
```

```python
# Basic statistics
print("Summary statistics:")
print(df_clean[['DAP', 'UREA', 'MOP']].describe())

# Visualize distributions
fig, axes = plt.subplots(2, 3, figsize=(15, 8))

# Histograms
for i, fertilizer in enumerate(['DAP', 'UREA', 'MOP']):
    axes[0, i].hist(df_clean[fertilizer], bins=15, edgecolor='black', alpha=0.7)
    axes[0, i].set_title(f'Distribution of {fertilizer} Prices')
    axes[0, i].set_xlabel('Price (NPR)')
    axes[0, i].set_ylabel('Frequency')

# Boxplots
for i, fertilizer in enumerate(['DAP', 'UREA', 'MOP']):
    axes[1, i].boxplot(df_clean[fertilizer])
    axes[1, i].set_title(f'Boxplot of {fertilizer} Prices')
    axes[1, i].set_ylabel('Price (NPR)')

plt.tight_layout()
plt.show()

# Check correlation
correlation_matrix = df_clean[['DAP', 'UREA', 'MOP']].corr()
print("\nCorrelation matrix:")
print(correlation_matrix)
```

```python
# Identify outliers using IQR method
def find_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

for fert in ['DAP', 'UREA', 'MOP']:
    outliers, low, high = find_outliers_iqr(df_clean, fert)
    print(f"\nOutliers in {fert} (below {low:.0f} or above {high:.0f}):")
    if not outliers.empty:
        print(outliers[['Zone', fert]])
    else:
        print("None")

# Check price consistency (UREA and MOP should be less than DAP)
df_clean['price_consistent'] = (df_clean['UREA'] < df_clean['DAP']) & (df_clean['MOP'] < df_clean['DAP'])
inconsistent = df_clean[df_clean['price_consistent'] == False]
print("\nZones with inconsistent pricing (UREA or MOP >= DAP):")
print(inconsistent[['Zone', 'DAP', 'UREA', 'MOP']])
```

```python
# 1. Identify most and least expensive zones
print("Top 5 most expensive zones (by DAP price):")
print(df_clean.nlargest(5, 'DAP')[['Zone', 'DAP', 'UREA', 'MOP']])

print("\nTop 5 least expensive zones (excluding Gulriya outlier):")
df_without_outlier = df_clean[df_clean['Zone'] != 'Gulriya']
print(df_without_outlier.nsmallest(5, 'DAP')[['Zone', 'DAP', 'UREA', 'MOP']])

# 2. Calculate price gaps
df_clean['DAP_UREA_gap'] = df_clean['DAP'] - df_clean['UREA']
df_clean['DAP_MOP_gap'] = df_clean['DAP'] - df_clean['MOP']

print("\nZones with largest price gap between DAP and UREA:")
print(df_clean.nlargest(5, 'DAP_UREA_gap')[['Zone', 'DAP', 'UREA', 'DAP_UREA_gap']])

# 3. Regional patterns (if we can infer regions from zone names)
# For example, group by hill/terai based on zone name patterns
def categorize_region(zone):
    hill_zones = ['Kathmandu', 'Dhulikhel', 'Dhankuta', 'Ilam', 'Palpa', 'Parbat', 'Doti']
    if zone in hill_zones:
        return 'Hill'
    elif 'gunj' in zone.lower() or 'nagar' in zone.lower() or 'ganj' in zone.lower():
        return 'Terai/Urban'
    else:
        return 'Other'

df_clean['Region'] = df_clean['Zone'].apply(categorize_region)
print("\nAverage prices by region:")
print(df_clean.groupby('Region')[['DAP', 'UREA', 'MOP']].mean().round(0))

# 4. Create a visualization of all zones
plt.figure(figsize=(14, 10))
zones_sorted = df_clean.sort_values('DAP', ascending=True)
plt.barh(zones_sorted['Zone'], zones_sorted['DAP'], color='skyblue', label='DAP')
plt.barh(zones_sorted['Zone'], zones_sorted['UREA'], color='lightgreen', alpha=0.7, label='UREA')
plt.barh(zones_sorted['Zone'], zones_sorted['MOP'], color='coral', alpha=0.5, label='MOP')
plt.xlabel('Price (NPR)')
plt.title('Fertilizer Prices Across Nepal Zones')
plt.legend()
plt.tight_layout()
plt.show()
```
